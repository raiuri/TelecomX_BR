```
import pandas as pd
```

```
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json"
```

```
df = pd.read_json(url)
```

```
df_normalizado = pd.json_normalize(df.to_dict(orient='records'));
```

```
df_normalizado
```

```
# limpeza dos dados
df_normalizado = df.drop_duplicates()

df_normalizado['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')

df_normalizado = df.dropna()
```

```
df_normalizado["Contas_Diarias"] = df["account.Charges.Monthly"] / 30
df_normalizado
```

```
mapeamento_binario = {
    "Yes": 1, "No": 0,
    "Sim": 1, "Não": 0,
    "Female": 0, "Male": 1
}

colunas_binarias = [
    "Churn", "customer.gender", "customer.Partner", "customer.Dependents",
    "phone.PhoneService", "phone.MultipleLines",
    "internet.OnlineSecurity", "internet.OnlineBackup", "internet.DeviceProtection",
    "internet.TechSupport", "internet.StreamingTV", "internet.StreamingMovies",
    "account.PaperlessBilling"
]

df_normalizado[colunas_binarias] = df_normalizado[colunas_binarias].replace(mapeamento_binario)
df_normalizado
```

```
descricao = df_normalizado.describe()
mediana = df_normalizado.median(numeric_only=True)
descricao.loc['median'] = mediana
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df_normalizado, x='Churn')
plt.title('Distribuição de Evasão de Clientes (Churn)')
plt.xlabel('Churn')
plt.ylabel('Número de Clientes')
plt.tight_layout()
plt.show()
```

```
churn_counts = df_normalizado['Churn'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Proporção de Clientes que Saíram vs Permaneceram')
plt.tight_layout()
plt.show()
```

```
categorical_cols = ['customer.gender', 'account.Contract', 'account.PaymentMethod']

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df_normalizado, x=col, hue='Churn', palette='Set2')
    plt.title(f'Distribuição de Churn por {col}')
    plt.xlabel(col)
    plt.ylabel('Número de Clientes')
    plt.legend(title='Churn', labels=['Não', 'Sim'])
    plt.tight_layout()
    plt.show()
```

```
numeric_cols = ['account.Charges.Total', 'customer.tenure', 'account.Charges.Monthly', 'Contas_Diarias']

for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_normalizado, x='Churn', y=col)
    plt.title(f'Distribuição de {col} por Churn')
    plt.xlabel('Churn')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

# Estatísticas descritivas agrupadas por churn
df_normalizado.groupby('Churn')[numeric_cols].describe().T
```
